### Hypothesis figure
Plotting hypothesis figure (figure 1) below

```python
# General summary figure 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches

plt.rcParams["font.family"] = "helvetica"
plt.rcParams["font.size"] = 16
plt.rcParams['lines.solid_capstyle'] = 'round'

conditions = [60,80]
colors = plt.cm.viridis(np.linspace(0, 0.45, len(conditions)))
titles = ['low cue validity','high cue validity']
cycles = 16

# Amplitude of the sine wave is sine of a variable like time
time = np.arange(0,cycles*np.pi,0.01)
sampling_rhythm   = np.sin(time) # 8 Hz

detection_performance = {}
detection_performance[60] = -np.sin(time*0.5+1.25*np.pi)
detection_performance[80] = 0.5*np.sin(time) + 0.5
index_1 = np.argmin(np.abs(time-0.5*np.pi))
index_2 = np.argmin(np.abs(time-4.5*np.pi))
index_3 = np.argmin(np.abs(time-8.5*np.pi))
index_4 = np.argmin(np.abs(time-12.5*np.pi))
detection_performance[80][index_1:index_2] = np.sin(time*0.5+0.25*np.pi)[index_1:index_2]
detection_performance[80][index_3:index_4] = np.sin(time*0.5+0.25*np.pi)[index_1:index_2]

# Plot a sine wave using time and amplitude obtained for the sine wave
fig = plt.figure(figsize=(12,6))
gs = fig.add_gridspec(2,3)

start_rect = 0
rect_width = 1
rect_height = 1

# rectangles
ax = {}
for axis in range(2):
    ax[axis] = fig.add_subplot(gs[axis,0])
    
    # add rectangles
    ax[axis].add_patch(plt.Rectangle((start_rect,0), rect_width, rect_height, facecolor = 'grey',edgecolor=colors[axis],linewidth=3,clip_on=False))
    
    # add text
    ax[axis].text(0.5,0.5,'>>•>>',va='center',ha='center',font='arial')
    ax[axis].text(0.5,0.6,'RVF cue',va='center',ha='center',size=11,fontweight='bold')
    
    # add title
    ax[axis].set_title(titles[axis],color = colors[axis],size=20,ha='center')
    
    # add ellipse and text left
    ax[axis].add_artist(patches.Ellipse((0.25,0.5), 0.2,0.28, facecolor = 'lightgrey',edgecolor = colors[axis], linestyle = 'dashed', linewidth=2,zorder = 1))
    ax[axis].text(0.25,0.5,str(100-conditions[axis])+'%',va='center',ha='center',size=18,fontweight='bold')
    ax[axis].text(0.25,0.15,'invalid',va='center',ha='center',size=18,fontweight='bold')
    
    # add ellipse and text right
    ax[axis].add_artist(patches.Ellipse((0.75,0.5), 0.2,0.28, facecolor = 'lightgrey',edgecolor = colors[axis], linewidth=2))
    ax[axis].text(0.75,0.5,str(conditions[axis])+'%',va='center',ha='center',size=18,fontweight='bold')
    ax[axis].text(0.75,0.15,'valid',va='center',ha='center',size=18,fontweight='bold')
    
    # delete lines
    ax[axis].axis('off')
    
# add time courses
ax = {}
for axis in range(len(conditions)):
    ax[axis] = fig.add_subplot(gs[axis,1:3])
    
    # plot sampling rhythm
    ax[axis].plot(time, sampling_rhythm, color = 'grey',linewidth = 2,alpha=0.5)
    
    # add text indicating lines
    ax[axis].text(time[-1]+0.5,0,'8 Hz\nsampling\nrhythm',size = 14,color='grey',va='center',ha='left')
    
    # add text indicating arrows (see below)
    ax[axis].text(time[-1]-2,1.15,'attentional\nshifts',size = 14,color='black',va='center',ha='left')
    
    # set x axis
    ax[axis].set_xlim(0,cycles*np.pi)
    ax[axis].tick_params(width=2)
    
    # set y axis
    ax[axis].set_ylim(-1,1)
    ax[axis].set_yticks([-1,0,1])
    ax[axis].set_yticklabels(['low','avg','high'])
    ax[axis].set_ylabel('RVF detection\nperformance')
    for x in ['top','right']:
        ax[axis].spines[x].set_linewidth(0)
    for x in ['left','bottom']:
        ax[axis].spines[x].set_linewidth(2)

# set x axis
ax[0].set_xticks([])
ax[1].set_xticks([0,cycles*np.pi])
ax[1].set_xticklabels(['0','1'])
ax[1].set_xlabel('time (s)',ha='center',va='center') # custom y label

# plot detection performance 60% and 80%
y = np.linspace(0.5*np.pi,(cycles-0.5)*np.pi,cycles)

# valid
ax[0].plot(time,detection_performance[60],color=colors[0],linewidth = 3,clip_on = False)
ax[1].plot(time,detection_performance[80],color=colors[1],linewidth = 3,clip_on = False)

# invalid
ax[0].plot(time,-detection_performance[60],color=colors[0],alpha=0.5,linestyle = 'dashed', linewidth = 3,clip_on = False)
ax[1].plot(time,-detection_performance[80],color=colors[1],alpha=0.5,linestyle = 'dashed', linewidth = 3,clip_on = False)

# plot attentional shift arrows 
arrow_x = np.linspace(0,time[-1],int(cycles/2)+1)
arrow_x = arrow_x[:-1]
arrow_length = time[-1] / cycles

# 60% attentional shift arrows
for i,x in enumerate(arrow_x):
    if i % 2 == 0: # if even, point arrow towards the (right)
        ax[0].arrow(x, 1.15, arrow_length, 0, head_width=0.1, head_length=1, length_includes_head = True, fc='k', ec='k',linewidth=2,clip_on = False) 
    elif i % 1 == 0:
        ax[0].arrow(x+arrow_length, 1.15, -arrow_length, 0, head_width=0.1, head_length=1, length_includes_head = True, fc='k', ec='k',linewidth=2,clip_on = False) 
    
# 80% attentional shift arrows
left_index = [1,5]
for i,x in enumerate(arrow_x):
    if i in left_index: # if even, point arrow towards the (right)
        ax[1].arrow(x+arrow_length, 1.15, -arrow_length, 0, head_width=0.1, head_length=1, length_includes_head = True, fc='k', ec='k',linewidth=2,clip_on = False) 
    else:
        ax[1].arrow(x, 1.15, arrow_length, 0, head_width=0.1, head_length=1, length_includes_head = True, fc='k', ec='k',linewidth=2,clip_on = False) 

fig.tight_layout(w_pad=1, h_pad=1.1)    

#plt.savefig(figure_folder+'figure_1.png',dpi=300)
```

<b>Figure 1</b>: Hypothesis figure. Schematic representation of detection performance of stimuli in the right visual field (RVF) over time, in conditions where the cue validity is low (60%) or high (80%). The solid lines represent detection performance for validly cued trials, the dashed lines represent detection performance for invalidly cued trials (‘avg’ = average). The grey line represents the underlying 8 Hz sampling rhythm or ‘clock’, providing periods of heightened or diminished perceptual sensitivity. In periods of heightened sensitivity, attentional shifts (represented by the black arrows) could occur towards either visual field. When the cue validity is low, there is a more balanced allocation of attentional resources, leading to as many attentional shifts towards the validly cued location in the RVF as the invalidly cued location in the LVF (left visual field). This results in a ~4 Hz rhythm of detection performance on the task over time. When the cue validity is high, however, the task relevance of the validly cued location is much higher than the task relevance of the invalid location. As a result, there are attentional shifts towards the validly cued RVF in almost every moment of heightened sensitivity. These shifts towards the validly cued location render reorienting from the valid towards the invalid location more challenging, resulting in a ~8Hz detection performance over time.
