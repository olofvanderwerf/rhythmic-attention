### Purpose of this notebook

This notebook visualises the outcomes of the stLSS analysis and calculates statistics.

### Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
import scipy.stats as stats
from scipy.signal import butter,lfilter
```

### Variables

```python
# folders
results_folder = 'data/results/'
figure_folder = 'figures/'

# time bins
start_time = 0.490
end_time = 1.690
resolution = 0.001
num_time_bins = int((end_time - start_time) / resolution) + 2
intervals = np.round(np.linspace(start_time, end_time, num = num_time_bins),3)

# frequency bins
low_freq = 2
high_freq = 20
resolution = 0.1
num_freq_bins = int((high_freq - low_freq) / resolution) + 1
frequencies =  np.linspace(low_freq, high_freq, num = num_freq_bins)

# Variables
validities = ['valid','invalid']
visual_fields = ['left','right','both']
conditions = ['60','80','100']

subjects = ['03','04','05','06','09','11','12','14','15','17','18','19','20','21','26','27','30','31','32','33','34','35','37','38','39','40']
```

### Import data
Import time series, observed power spectra and surrogate power spectra 

```python
# Import time series
time_series = {}

for validity in validities:
    time_series[validity] = {}
        
    for visual_field in visual_fields:
        time_series[validity][visual_field] = {}
        
        for condition in conditions:  
            if validity == 'invalid' and condition == '100':
                continue
                
            time_series_file = results_folder +'time_series/'+validity+'_'+visual_field+'_'+condition+'.csv'
            ts = pd.read_csv(time_series_file, sep = ',',index_col=0).squeeze("columns")
            
            # detrend
            detrended = signal.detrend(ts, axis=0, type='linear')
            detrended = pd.Series(detrended, index = ts.index)
            
            # Add to time series dictionary
            time_series[validity][visual_field][condition] = detrended
```

```python
# Import power spectra
power_spectrum = {}
for validity in validities:
    power_spectrum[validity] = {}
    for visual_field in visual_fields:
        power_spectrum[validity][visual_field] = {}
        for condition in conditions:
            if validity == 'invalid' and condition == '100':
                continue
                
            file = results_folder + 'power_spectra/' + validity + '_' + visual_field + '_' + condition + '.csv'
            power_spectrum[validity][visual_field][condition] = pd.read_csv(file,index_col = 0)
```

There is a sampling artifact in the power spectrum in the 80% cue validity condition, for invalid trials, when left and right visual fields separately:

```python
# plot power spectra
plt.plot(power_spectrum['invalid']['left']['80'])
plt.title('Power spectrum, invalid trials, left visual field, 80% cue validity:')
plt.show()
plt.plot(power_spectrum['invalid']['right']['80'])
plt.title('Power spectrum, invalid trials, right visual field, 80% cue validity:')
plt.show()
```

Thus, we will numb out the data (put in 'NaN') at 15.0 Hz, so as to not disturb plotting and comparison against the surrogate power spectra.

```python
# make sure that 15 Hz artifact is numbed out
power_spectrum['invalid']['left']['80'].loc[15.0] = np.nan
power_spectrum['invalid']['right']['80'].loc[15.0] = np.nan
```

```python
# Import surrogate power spectra - interval labels
surrogate_power_spectrum = {}
for validity in validities:
    surrogate_power_spectrum[validity] = {}
    for visual_field in visual_fields:
        surrogate_power_spectrum[validity][visual_field] = {}
        for condition in conditions:
            if validity == 'invalid' and condition == '100':
                continue
                
            file = results_folder + 'permutations/interval_permutations/' + validity + '_' + visual_field + '_' + condition +'.csv'
            surrogate_power_spectrum[validity][visual_field][condition] = pd.read_csv(file, index_col = 0)
```

Same for the surrogate power spectrum: we will numb out the data (put in 'NaN') at 15.0 Hz, so as to not disturb plotting and comparison with the observed power spectrum.

```python
# make sure that 15 Hz artifact is numbed out
surrogate_power_spectrum['invalid']['left']['80'].loc[15.0] = np.nan
surrogate_power_spectrum['invalid']['right']['80'].loc[15.0] = np.nan
```

```python
# Import surrogate power spectra - condition labels
surrogate_power_spectrum_conditions = {}
for visual_field in visual_fields:
    surrogate_power_spectrum_conditions[visual_field] = {}
    for condition in conditions:

        file = results_folder + 'permutations/condition_permutations/valid_' + visual_field + '_' + condition +'.csv'
        surrogate_power_spectrum_conditions[visual_field][condition] = pd.read_csv(file, index_col = 0)
```

### Calculate statistics
Calculating statistics: comparing observed power spectra against the surrogate power spectra, where cue-target interval bin labels are shuffled

```python
# Calculate statistics - interval label permutations
for validity in validities:
    for visual_field in visual_fields:
        for condition in conditions:
            
            if validity == 'invalid' and condition == '100':
                continue
                
            print('For '+validity+', '+visual_field+', '+condition+': ')
            # peak at...
            peak_value = power_spectrum[validity][visual_field][condition].max()[0]
            freq = power_spectrum[validity][visual_field][condition].idxmax()[0]
            print('Peak of '+str(peak_value)+' at '+str(freq)+' Hz.')
            non_corrected = 1 - (stats.percentileofscore(surrogate_power_spectrum[validity][visual_field][condition].loc[freq],peak_value)/100)
            print('Non corrected p:'+str(non_corrected))
            max_based = 1 - (stats.percentileofscore(surrogate_power_spectrum[validity][visual_field][condition].max(axis = 0),peak_value)/100)
            print('Max-based corrected p:'+str(max_based))
            pval = pd.Series(index=frequencies,dtype='float64')
            for freq in power_spectrum[validity][visual_field][condition].index:
                pval.loc[freq] = 1 - stats.percentileofscore(surrogate_power_spectrum[validity][visual_field][condition].max(axis = 0),power_spectrum[validity][visual_field][condition].loc[freq][0])/100

            print('Significant peak frequencies:')
            print(list(pval[pval <= 0.05].index))
            
```

Comparing observed power spectra against the surrogate power spectra, where condition labels are shuffled

```python
# Calculate statistics - condition label permutations
for visual_field in visual_fields:
    for condition in conditions:
        print('For '+visual_field+', '+condition+':')
        # peak at...
        peak_value = power_spectrum['valid'][visual_field][condition].max()[0]
        freq = power_spectrum['valid'][visual_field][condition].idxmax()[0]
        print('Peak of '+str(peak_value)+' at '+str(freq)+' Hz.')
        non_corrected = 1 - (stats.percentileofscore(surrogate_power_spectrum_conditions[visual_field][condition].loc[freq],peak_value)/100)
        print('Non corrected:'+str(non_corrected))
        max_based = 1 - (stats.percentileofscore(surrogate_power_spectrum_conditions[visual_field][condition].max(axis = 0),peak_value)/100)
        print('Max-based:'+str(max_based))
        pval = pd.Series(index=power_spectrum['valid'][visual_field][condition].index,dtype='float64')
        for freq in power_spectrum['valid'][visual_field][condition].index:
            pval.loc[freq] = 1 - stats.percentileofscore(surrogate_power_spectrum_conditions[visual_field][condition].max(axis = 0),power_spectrum['valid'][visual_field][condition].loc[freq][0])/100

        print('Significant peak frequencies:')
        print(list(pval[pval <= 0.05].index))
```

### Summary figure
Plotting summary figure (figure 4) below

```python
# General summary figure 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches

plt.rcParams["font.family"] = "helvetica"
plt.rcParams["font.size"] = 16
plt.rcParams['lines.solid_capstyle'] = 'round'

conditions = [60,80]
colors = plt.cm.viridis(np.linspace(0, 0.45, len(conditions)))
titles = ['low cue validity','high cue validity']
cycles = 16

# Amplitude of the sine wave is sine of a variable like time
time = np.arange(0,cycles*np.pi,0.01)
sampling_rhythm   = np.sin(time) # 8 Hz

detection_performance = {}
detection_performance[60] = -np.sin(time*0.5+1.25*np.pi)
detection_performance[80] = 0.5*np.sin(time) + 0.5
index_1 = np.argmin(np.abs(time-0.5*np.pi))
index_2 = np.argmin(np.abs(time-4.5*np.pi))
index_3 = np.argmin(np.abs(time-8.5*np.pi))
index_4 = np.argmin(np.abs(time-12.5*np.pi))
detection_performance[80][index_1:index_2] = np.sin(time*0.5+0.25*np.pi)[index_1:index_2]
detection_performance[80][index_3:index_4] = np.sin(time*0.5+0.25*np.pi)[index_1:index_2]

# Plot a sine wave using time and amplitude obtained for the sine wave
fig = plt.figure(figsize=(12,6))
gs = fig.add_gridspec(2,3)

start_rect = 0
rect_width = 1
rect_height = 1

# rectangles
ax = {}
for axis in range(2):
    ax[axis] = fig.add_subplot(gs[axis,0])
    
    # add rectangles
    ax[axis].add_patch(plt.Rectangle((start_rect,0), rect_width, rect_height, facecolor = 'grey',edgecolor=colors[axis],linewidth=3,clip_on=False))
    
    # add text
    ax[axis].text(0.5,0.5,'>>>>',va='center',ha='center',font='arial')
    ax[axis].text(0.5,0.6,'RVF cue',va='center',ha='center',size=11,fontweight='bold')
    
    # add title
    ax[axis].set_title(titles[axis],color = colors[axis],size=20,ha='center')
    
    # add ellipse and text left
    ax[axis].add_artist(patches.Ellipse((0.25,0.5), 0.2,0.28, facecolor = 'lightgrey',edgecolor = colors[axis], linestyle = 'dashed', linewidth=2,zorder = 1))
    ax[axis].text(0.25,0.5,str(100-conditions[axis])+'%',va='center',ha='center',size=18,fontweight='bold')
    ax[axis].text(0.25,0.15,'invalid',va='center',ha='center',size=18,fontweight='bold')
    
    # add ellipse and text right
    ax[axis].add_artist(patches.Ellipse((0.75,0.5), 0.2,0.28, facecolor = 'lightgrey',edgecolor = colors[axis], linewidth=2))
    ax[axis].text(0.75,0.5,str(conditions[axis])+'%',va='center',ha='center',size=18,fontweight='bold')
    ax[axis].text(0.75,0.15,'valid',va='center',ha='center',size=18,fontweight='bold')
    
    # delete lines
    ax[axis].axis('off')
    
# add time courses
ax = {}
for axis in range(len(conditions)):
    ax[axis] = fig.add_subplot(gs[axis,1:3])
    
    # plot sampling rhythm
    ax[axis].plot(time, sampling_rhythm, color = 'grey',linewidth = 2,alpha=0.5)
    
    # add text indicating lines
    ax[axis].text(time[-1]+0.5,0,'8 Hz\nsampling\nrhythm',size = 14,color='grey',va='center',ha='left')
    
    # add text indicating arrows (see below)
    ax[axis].text(time[-1]-2,1.15,'attentional\nshifts',size = 14,color='black',va='center',ha='left')
    
    # set x axis
    ax[axis].set_xlim(0,cycles*np.pi)
    ax[axis].tick_params(width=2)
    
    # set y axis
    ax[axis].set_ylim(-1,1)
    ax[axis].set_yticks([-1,0,1])
    ax[axis].set_yticklabels(['low','avg','high'])
    ax[axis].set_ylabel('RVF detection\nperformance')
    for x in ['top','right']:
        ax[axis].spines[x].set_linewidth(0)
    for x in ['left','bottom']:
        ax[axis].spines[x].set_linewidth(2)

# set x axis
ax[0].set_xticks([])
ax[1].set_xticks([0,cycles*np.pi])
ax[1].set_xticklabels(['0','1'])
ax[1].set_xlabel('time (s)',ha='center',va='center') # custom y label

# plot detection performance 60% and 80%
y = np.linspace(0.5*np.pi,(cycles-0.5)*np.pi,cycles)

# valid
ax[0].plot(time,detection_performance[60],color=colors[0],linewidth = 3,clip_on = False)
ax[1].plot(time,detection_performance[80],color=colors[1],linewidth = 3,clip_on = False)

# invalid
ax[0].plot(time,-detection_performance[60],color=colors[0],alpha=0.5,linestyle = 'dashed', linewidth = 3,clip_on = False)
ax[1].plot(time,-detection_performance[80],color=colors[1],alpha=0.5,linestyle = 'dashed', linewidth = 3,clip_on = False)

# plot attentional shift arrows 
arrow_x = np.linspace(0,time[-1],int(cycles/2)+1)
arrow_x = arrow_x[:-1]
arrow_length = time[-1] / cycles

# 60% attentional shift arrows
for i,x in enumerate(arrow_x):
    if i % 2 == 0: # if even, point arrow towards the (right)
        ax[0].arrow(x, 1.15, arrow_length, 0, head_width=0.1, head_length=1, length_includes_head = True, fc='k', ec='k',linewidth=2,clip_on = False) 
    elif i % 1 == 0:
        ax[0].arrow(x+arrow_length, 1.15, -arrow_length, 0, head_width=0.1, head_length=1, length_includes_head = True, fc='k', ec='k',linewidth=2,clip_on = False) 
    
# 80% attentional shift arrows
left_index = [1,5]
for i,x in enumerate(arrow_x):
    if i in left_index: # if even, point arrow towards the (right)
        ax[1].arrow(x+arrow_length, 1.15, -arrow_length, 0, head_width=0.1, head_length=1, length_includes_head = True, fc='k', ec='k',linewidth=2,clip_on = False) 
    else:
        ax[1].arrow(x, 1.15, arrow_length, 0, head_width=0.1, head_length=1, length_includes_head = True, fc='k', ec='k',linewidth=2,clip_on = False) 

fig.tight_layout(w_pad=1, h_pad=1.1)    

plt.savefig(figure_folder+'figure_4.png',dpi=300)
```

<b>Figure 4</b>: Summary figure. Schematic representation of detection performance of stimuli in the right visual field (RVF) over time, in conditions where the cue validity is low (60%) or high (80%). The solid lines represent detection performance for validly cued trials, the dashed lines represent detection performance for invalidly cued trials (avg = average). The grey line represents the underlying 8 Hz sampling rhythm or clock, providing periods of heightened or diminished perceptual sensitivity. In periods of heightened sensitivity, attentional shifts (represented by the black arrows) could occur towards either visual field. When the cue validity is low, there is a more balanced allocation of attentional resources, leading to as many attentional shifts towards the validly cued location in the RVF as the invalidly cued location in the LVF (left visual field). This results in a ~4 Hz rhythm of detection performance on the task over time. When the cue validity is high, however, the task relevance of the validly cued location is much higher than the task relevance of the invalid location. As a result, there are attentional shifts towards the validly cued RVF in almost every moment of heightened sensitivity. These shifts towards the validly cued location render reorienting from the valid towards the invalid location more challenging, resulting in a ~8Hz detection performance over time.
